**HIERARCHIAL CLUSTERING**

**Example for Agglomerative Clustering**

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np

X=np.array([[1,2],[1,4],[1,0],[4,2],[4,4],[4,0]])
clustering=AgglomerativeClustering(n_clusters=2).fit(X)
print(clustering.labels_)
```

**Example for Divisive Clustering**

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram,linkage
import matplotlib.pyplot as plt

X=np.array([[1,2],[1,4],[1,0],[4,2],[4,4],[4,0]])

Z=linkage(X,'ward')
dendrogram(Z)

plt.title("Hierarchial Clustering Dendogram")
plt.xlabel('Data point')
plt.ylabel('Distance')
plt.show()
```

```python
# Strip spaces and fix column names
df.columns = df.columns.str.strip()

# Print columns to verify
print(df.columns)
```

```python
# Identify categorical columns
categorical_cols = ['Education', 'Marital_Status']

# Convert categorical columns to numeric using Label Encoding
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])  # Convert text to numbers
    label_encoders[col] = le  # Store encoders if you need to reverse later

# Now try scaling and clustering again
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df.drop(columns=['ID', 'Dt_Customer']))  # Exclude non-numeric columns
```

```python
from scipy.spatial.distance import pdist, squareform

distance_matrix = squareform(pdist(scaled_data, metric='euclidean'))
```

**Agglomerative Clustering & Dendrogram**

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
linkage_matrix = sch.linkage(scaled_data, method='ward')  # Agglomerative (bottom-up)
sch.dendrogram(linkage_matrix)
plt.title("Agglomerative Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

**Divisive Clustering & Dendrogram**

```python
plt.figure(figsize=(12, 6))
linkage_matrix_div = sch.linkage(scaled_data, method='ward', optimal_ordering=True)  # Top-down
sch.dendrogram(linkage_matrix_div)
plt.title("Divisive Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Fill missing values in 'Income' with mean
df['Income'].fillna(df['Income'].mean(), inplace=True)

# Select only numeric columns for clustering
X = df.select_dtypes(include=[np.number]).values  

# Scale the data for better clustering performance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# List to store Within-Cluster Sum of Squares (WCSS)
wcss = []

# Test K values from 1 to 10
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)  # Append WCSS value

# Plot the Elbow Graph
plt.figure(figsize=(8, 5))
plt.plot(K_range, wcss, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal k')
plt.show()

# Apply K-Means using the optimal k (manually chosen from the graph)
optimal_k = 3  # Change this based on elbow graph
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans_final.fit_predict(X_scaled)  # Assign clusters

# Display the first few rows with cluster labels
df.head()
```
