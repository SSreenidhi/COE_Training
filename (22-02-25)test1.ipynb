```python
# Define a function to find outliers based on IQR for each column
def find_outliers(df):
    outliers = pd.DataFrame()
    for col in df.select_dtypes(include=['float64', 'int64']).columns:  # Apply to numerical columns
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Get the rows where values are outside the bounds
        column_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        
        if not column_outliers.empty:
            outliers[col] = column_outliers[col]
    
    return outliers

# Find outliers for the entire dataset
outliers_in_dataset = find_outliers(data)

# Show outliers (if any)
print("Outliers in the entire dataset:")
outliers_in_dataset
```

```python
from scipy.stats import zscore

# Function to find outliers using Z-score for the entire dataset
def find_outliers_zscore(df, threshold=3):
    outliers = pd.DataFrame()
    for col in df.select_dtypes(include=['float64', 'int64']).columns:  # Apply to numerical columns
        z_scores = zscore(df[col])  # Calculate Z-scores
        column_outliers = df[abs(z_scores) > threshold]  # Identify outliers where Z-score > threshold
        
        if not column_outliers.empty:
            outliers[col] = column_outliers[col]  # Store outliers in the dataframe
    
    return outliers

# Find outliers using Z-score
outliers_zscore = find_outliers_zscore(data)

# Show outliers (if any)
print("Outliers based on Z-score:")
outliers_zscore
```

```python
# Calculate Z-scores manually
mean_stock_level = data['Stock_Level'].mean()
std_stock_level = data['Stock_Level'].std()

z_scores = (data['Stock_Level'] - mean_stock_level) / std_stock_level

# Define a threshold for outliers (commonly 3)
threshold = 3

# Identify rows where the absolute Z-score is greater than the threshold
outliers = data[abs(z_scores) > threshold]

# Display the outliers in the 'Stock_Level' column
print("Outliers in Stock_Level based on Z-score:")
print(outliers[['Stock_Level']])
```
