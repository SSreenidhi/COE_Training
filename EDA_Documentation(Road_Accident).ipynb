# **Outlier Detection and Removal**

(a) Using IQR Method

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Compute IQR
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Detect outliers
outliers = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).sum()
print("\n--- Outliers detected using IQR ---")
print(outliers)

# Visualizing outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=numeric_df)
plt.xticks(rotation=90)
plt.title("Boxplot for Outlier Detection")
plt.show()
```

# **Removing outliers and graph after removing the ouliers**

```python
# Removing Outliers using IQR
df_cleaned = df[~((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Visualizing data after removing outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_cleaned.select_dtypes(include=[np.number]))
plt.xticks(rotation=90)
plt.title("Boxplot After Removing Outliers")
plt.show()
```

# **(b) Using Z-score Method**

```python
from scipy import stats
z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))
outliers_z = (z_scores > 3).sum()
print("\n--- Outliers detected using Z-score ---")
print(outliers_z)

# Removing Outliers using Z-score
df = df[(z_scores < 3).all(axis=1)]

# Visualizing data after removing Z-score outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Boxplot After Removing Z-score Outliers")
plt.show()
```

# **Probability Distributions (Binomial, Poisson, Normal, etc.)**
📌 Why?

Helps understand the spread & nature of accident data
Identifies if the data follows normal, binomial, or Poisson distribution

🔹 Which Distribution to Use?

✔ Binomial Distribution → If data is binary (e.g., accident occurred or not)

✔ Poisson Distribution → If counting rare events (e.g., accidents per day)

✔ Normal Distribution → If data is symmetrically distributed

```python
import scipy.stats as stats

# Histogram & Normal Distribution Fit
sns.histplot(df['Accidents_2019'], kde=True, bins=20)
plt.title("Accidents in 2019 - Distribution")
plt.show()

# Check normality
stat, p = stats.shapiro(df['Accidents_2019'])
print("Shapiro-Wilk Test p-value:", p)  # If p < 0.05, data is NOT normal

# Poisson Distribution Fit
sns.histplot(np.random.poisson(lam=df['Accidents_2019'].mean(), size=1000), kde=True, color='r')
plt.title("Poisson Distribution Fit")
plt.show()
```

# **Diamentional Reduction**
# **Principal Component Analysis (PCA)**
✔ Reduces correlated features

✔ Retains maximum variance in fewer components

✔ Helps in faster training of ML models

```python
df.dropna(inplace=True)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Selecting only numerical columns for PCA
numerical_cols = df.select_dtypes(include=['number']).columns
df_num = df[numerical_cols]

# Standardizing data (PCA works better with scaled data)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_num)

# Applying PCA (reduce to 2D for visualization)
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)

# Adding PCA components to the DataFrame
df['PCA_1'] = pca_result[:, 0]
df['PCA_2'] = pca_result[:, 1]

# Scatter plot of PCA components
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['PCA_1'], y=df['PCA_2'], hue=df['Rank_2019'], palette='viridis')
plt.title("PCA - 2D Representation of Road Accidents Data")
plt.show()

# Explained Variance
print("Explained Variance by Components:", pca.explained_variance_ratio_)
```
