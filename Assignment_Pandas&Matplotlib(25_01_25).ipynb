# **Sales Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt
sales_df=pd.read_csv("/content/sales_data.csv")
sales_df.head()
```

#Line Graph

```python
plt.figure(figsize=(10, 6))
plt.plot(sales_df['Month'], sales_df['Sales'], marker='o', color='b', linestyle='-')
plt.title('Monthly Sales Analysis')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

#Histogram

```python
plt.figure(figsize=(10, 6))
plt.hist(sales_df['Order_Quantity'], bins=10, color='green', edgecolor='black', alpha=0.7)
plt.title('Order Quantity Distribution')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **Customer Demographics**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

cust_df = pd.read_csv("/content/customer_demographics.csv")
cust_df.head()
```

#Histogram

```python
plt.figure(figsize=(10, 6))
plt.hist(cust_df['Age'], bins=10, color='purple', edgecolor='black', alpha=0.7)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Scatter plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(cust_df['Annual_Income'], cust_df['Spending_Score'], color='darkblue', alpha=0.7)
plt.title('Spending Patterns vs Annual Income')
plt.xlabel('Annual Income ')
plt.ylabel('Spending Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# **School Performance**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

school_df= pd.read_csv("School_Performance.csv")
school_df.head()
```

# BAR CHAT

```python
sub = ["Math", "Science", "English", "History", "Geography"]
sub_means =school_df[sub].mean()

plt.figure(figsize=(10, 6))
sub_means.plot(kind='bar', color='orange', edgecolor='black')
plt.title("Average Marks Across Subjects")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#Box plot

```python
plt.figure(figsize=(10, 6))
sub = ["Math", "Science", "English", "History", "Geography"]
sns.boxplot(data=school_df[sub], palette="flare")
plt.title("Box Plot of Marks Across Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(axis='y', alpha=0.7)
plt.tight_layout()
plt.show()
```
