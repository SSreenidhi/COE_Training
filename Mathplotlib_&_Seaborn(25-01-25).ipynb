```python
import pandas as pd
import matplotlib.pyplot as plt

data['dt'] = pd.to_datetime(data['dt'])
country_avg_temp = data.groupby('Country')['AverageTemperature'].mean().reset_index()

country_avg_temp = country_avg_temp.sort_values(by='AverageTemperature', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(country_avg_temp['Country'], country_avg_temp['AverageTemperature'], color='skyblue')
plt.title('Countries with the Highest Average Temperature')
plt.xlabel('Country')
plt.ylabel('Average Temperature (°C)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

# BAR PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt

country_avg_temp = data.groupby('Country')['AverageTemperature'].mean().reset_index()

country_avg_temp = country_avg_temp.sort_values(by='AverageTemperature', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(country_avg_temp['Country'], country_avg_temp['AverageTemperature'], color='pink')
plt.title('Top 10 Countries with the Highest Average Temperature')
plt.xlabel('Country')
plt.ylabel('Average Temperature (°C)')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

```

# HISTOGRAM 

```python
plt.figure(figsize=(10, 6))
plt.hist(data['AverageTemperature'].dropna(), bins=50, color='lightseagreen', edgecolor='black')
plt.title('Average Temperature across all dates')
plt.xlabel('Temperature Ranges')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

# SCATTER PLOT

```python
plt.figure(figsize=(10,6))
data.dropna()
plt.scatter(data['AverageTemperature'],data['AverageTemperatureUncertainty'], color='purple')
plt.title('Corecllation between temperature and uncertainity')
plt.xlabel('Average Temperature')
plt.ylabel('Average Temperature Uncertainity')
plt.show()
```

# BOX PLOT|

```python
import matplotlib.pyplot as plt

# Filter the dataset for the specific country
country_data = data[data['Country'] == 'India'].dropna()

# Extract the relevant data: Year and AverageTemperature
country_data['Year'] = pd.to_datetime(country_data['dt']).dt.year.head(100)
temperature_data = country_data[['Year', 'AverageTemperature']]

# Plot boxplot for temperature spread by year
plt.figure(figsize=(12, 6))
plt.boxplot(
    [temperature_data[temperature_data['Year'] == year]['AverageTemperature'] for year in temperature_data['Year'].unique()],
    labels=temperature_data['Year'].unique(),
    patch_artist=True,
    boxprops=dict(facecolor='pink', color='black'),
    medianprops=dict(color='darkblue', linewidth=1.5),
    flierprops=dict(marker='o', color='red', alpha=0.5),
)

plt.title('Spread of Average Temperatures in the India (by Year)', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Temperature', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color="blue")
plt.title("Distribution of the Total Bills")
plt.show()
```

# KDE PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data=tips, x="total_bill", fill=True, color="crimson")
plt.title("KDE of the Total Bills")
plt.show()
```

# BAR PLOT

```python
sns.barplot(data=tips,x="day" ,y="total_bill",palette="viridis")
plt.title("Average Total Bill By A Day")
plt.show()
```

# BOX PLOT

```python
sns.boxplot(data=tips,x="day" ,y="total_bill",hue="sex",palette="coolwarm")
plt.title("Box Plot of Total Bills by Day and Gender")
plt.show()
```

# SCATTER PLOT

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",palette="deep")
plt.title("Scatter plot of Total Bills vs.Tip")
plt.show()
```

# LINE PLOT

```python
sns.lineplot(data=tips,x="size",y="tip",marker="o",color="purple")
plt.title("Line size of Party Size vs.Tip")
plt.show()
```

# PAIR PLOT

```python
sns.pairplot(data=tips,hue="sex",palette="husl")
plt.show()
```

# HEATMAP

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_data = tips.select_dtypes(include=['number'])
corr_matrix = numeric_data.corr()

sns.heatmap(corr_matrix, annot=True, cmap="flare", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# SEABORN ON TITANIC DATASET

```python
import pandas as pd
train_df=pd.read_csv("train.csv")
train_df.head()
```

# BAR PLOT 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(data=train_df,x='Pclass',y='Survived',errorbar=None)
plt.title('Survival Rate by Passenger Id')
plt.xlabel('Passenger class')
plt.ylabel('Survival Rate')
plt.show()
```

# SCATTER PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=train_df, x="Age", y="Fare", hue="Survived",palette="husl")
plt.title("Scatter plot of Age vs Fare")
plt.show()
```

# PAIR PLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(data=train_df[["Age", "Fare", "Pclass","Survived"]], hue="Survived", palette="magma")
plt.show()
```

# HEAT MAP

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_data =train_df.select_dtypes(include=['number'])
corr_matrix = numeric_data.corr()

sns.heatmap(corr_matrix, annot=True, cmap="flare", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
