# **Assignment on Uber dataset**

**Load the dataset**

```python
import pandas as pd
uber_data=pd.read_csv("Uber.csv")
uber_data.head()
```

**Display basic info about dataset**

```python
print(uber_data.info())
print(uber_data.describe())
```

**Check for missing values**

```python
print(uber_data.isnull().sum())
```

**Drop rows with missing values**

```python
uber_cleaned_data = uber_data.dropna(how='all')
uber_cleaned_data
```

**fill missing values (propose column with unknown value)**

```python
uber_data = uber_data.fillna('Unknown')
uber_data
```

**Check and remove duplicates**

```python
#to check duplicate
duplicates = uber_data[uber_data.duplicated()]
duplicates
```

```python
uber_data = uber_data.drop_duplicates()
uber_data
```

**Convert START_DATE and END_DATE to datetime**

```python
uber_data['START_DATE*'] = pd.to_datetime(uber_data['START_DATE*'], errors='coerce')
uber_data['END_DATE*'] = pd.to_datetime(uber_data['END_DATE*'], errors='coerce')
uber_data.dtypes
```

**Total number of rides per category:**

```python
total_rides = uber_data['CATEGORY*'].value_counts()
total_rides
```

**Total miles traveled for each purpose:**

```python
total_miles_per_purpose = uber_data.groupby('PURPOSE*')['MILES*'].sum()
total_miles_per_purpose
```

**Average distance for business vs. personal rides:**

```python
avg_dis= uber_data.groupby('CATEGORY*')['MILES*'].mean()
avg_dis
```

**Add a column for cost estimation (assuming $2 per mile):**

```python
uber_data['COST_ESTIMATION*'] = uber_data['MILES*'] * 2
#uber_data.loc[:, 'COST_ESTIMATION*'] = uber_data['MILES*'] * 2
uber_data.head()
```

**Filter rides longer than 50 miles:**

```python
d = uber_data[uber_data['MILES*'] >= 50][['MILES*', 'START*', 'STOP*']]
d
```

**Filter by specific purpose (e.g., meetings):**

```python
meetings= uber_data[uber_data['PURPOSE*'] == 'Meeting']
meetings
```

**What is the total number of business trips versus personal trips?**

```python
total_trips= uber_data['CATEGORY*'].value_counts()
print(total_trips)
```

**What percentage of trips are business versus personal?**

```python
count_of_categeory = uber_data['CATEGORY*'].value_counts()
category_percent = (count_of_categeory / len(uber_data)) * 100
print(category_percent)
```
